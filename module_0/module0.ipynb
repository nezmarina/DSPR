```python
import numpy as np
"""Программа загадывает число в диапазоне от 1 до 100"""
number = np.random.randint(1,101)
print ("Загадано число от 1 до 100")

def game_core(number):
    """Определяем функцию по поиску загаданного числа"""    
    count=0
    predict=50
    left_position=1
    right_position=100
    
    while predict!=number:
        count+=1
        
        if number>predict:            
            left_position = predict + 1
            predict = (left_position + right_position)//2
            
        else:
            right_position = predict - 1
            predict = (left_position + right_position)//2
            
    count+=1
    print (f"Вы угадали число {number} за {count} попыток.")
    return(count)     


def score_game(game_core):
    '''Запускаем игру 1000 раз, чтобы узнать, как быстро игра угадывает число'''
    count_ls = []
    np.random.seed(1)
    random_array = np.random.randint(1,101, size=(1000))
    
    for number in random_array:
        count_ls.append(game_core(number))
        
    score = int(np.mean(count_ls))
    print(f"Ваш алгоритм угадывает число в среднем за {score} попыток")
    return(score)

print(score_game(game_core))
```
